```python
import threading
import random
import time

def gerador_numeros_aleatorios(thread_id):
    while True:
        numero_aleatorio = random.randint(1, 100)
        print(f'Thread {thread_id}: {numero_aleatorio}')
        time.sleep(1)

threads = []

for i in range(5):
    thread = threading.Thread(target=gerador_numeros_aleatorios, args=(i,))
    threads.append(thread)
    thread.start()

for thread in threads:
    thread.join()
```
